# Fase 2: EDA (Análisis Exploratorio de Datos)
Exploraremos el dataset limpio para encontrar patrones, distribuciones y correlaciones.

### 1. Cargar Datos y Cabecera

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración visual de seaborn
sns.set_theme(style='whitegrid')

# Cargar los datos limpios
df_limpio = pd.read_csv('EDUCACION_limpio.csv')

# Mostrar la cabecera
df_limpio.head()
```

### 2. Información General del Dataset
Visualizamos los tipos de datos y estadísticas descriptivas para entender la magnitud de las variables.

```
# Información de tipos de datos
print('INFORMACIÓN DEL DATASET:')
df_limpio.info()

# Estadísticas descriptivas
print('\nESTADÍSTICAS DESCRIPTIVAS:')
df_limpio.describe()
```

### 3. Distribución de la Tasa de Aprobación
Veamos cómo se distribuye nuestra variable objetivo.

```
plt.figure(figsize=(10, 6))
sns.histplot(df_limpio['APROBACIÓN'], bins=30, kde=True, color='blue')
plt.title('Distribución de la Tasa de Aprobación')
plt.xlabel('Tasa de Aprobación (%)')
plt.ylabel('Frecuencia')
plt.show()
```

### 4. Matriz de Correlación
Revisamos cómo se relacionan las variables de cobertura y población con la aprobación y deserción.

```
cols_numericas = ['APROBACIÓN', 'DESERCIÓN', 'REPROBACIÓN', 'COBERTURA_NETA', 'COBERTURA_BRUTA', 'POBLACIÓN_5_16']

plt.figure(figsize=(10, 8))
correlacion = df_limpio[cols_numericas].corr()
sns.heatmap(correlacion, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Mapa de Calor de Correlaciones')
plt.show()
```
