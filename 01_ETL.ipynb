# Fase 1: ETL (Extracción, Transformación y Carga)
En este cuaderno realizaremos la limpieza profunda del dataset original `EDUCACION.csv`.

### 1. Extracción de Datos
Cargamos la base de datos original.

```
import pandas as pd
import numpy as np

# Cargamos el dataset
# Utilizamos el encoding adecuado si es necesario (ej. utf-8 o latin1)
df = pd.read_csv('EDUCACION.csv')
print('Dimensiones del dataset:', df.shape)
```

### 2. Transformación de Datos
Muchas variables vienen en formato de texto con el símbolo `%` y comas como separador decimal. Procedemos a limpiarlos y transformarlos a numéricos.

```
# Funciones de limpieza
def clean_percentage(val):
    if pd.isna(val):
        return np.nan
    if isinstance(val, str):
        return float(val.replace('%', '').replace('.', '').replace(',', '.'))
    return val

def clean_population(val):
    if pd.isna(val):
        return np.nan
    if isinstance(val, str):
        return float(val.replace(',', '').replace('.', ''))
    return val

# Columnas de porcentajes
cols_percentage = [
    'APROBACIÓN', 'DESERCIÓN', 'REPROBACIÓN', 
    'COBERTURA_NETA', 'COBERTURA_BRUTA'
]

# Limpiar porcentajes
for col in cols_percentage:
    if col in df.columns:
        df[col] = df[col].apply(clean_percentage)

# Limpiar población
if 'POBLACIÓN_5_16' in df.columns:
    df['POBLACIÓN_5_16'] = df['POBLACIÓN_5_16'].apply(clean_population)

cols_all = cols_percentage + ['POBLACIÓN_5_16']
df[cols_all].head()
```

### 3. Manejo de Valores Nulos
Imputamos los valores nulos utilizando la mediana para no afectar la distribución general por valores atípicos.

```
# Rellenamos los valores nulos con la mediana de cada columna
df[cols_to_clean] = df[cols_to_clean].fillna(df[cols_to_clean].median())

# Verificamos si quedan nulos
print('Valores nulos restantes:\n', df[cols_to_clean].isnull().sum())
```

### 4. Carga (Guardar el Dataset Limpio)
Exportamos el dataset limpio para que sea utilizado en la siguiente fase (EDA).

```
# Guardamos el nuevo archivo CSV limpio
df.to_csv('EDUCACION_limpio.csv', index=False)
print('Dataset guardado exitosamente como EDUCACION_limpio.csv')
```
